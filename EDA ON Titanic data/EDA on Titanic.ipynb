```python
# Feature enginerring 
# we create one new column family size= parch+sisb
df['Family_size']=df['Parch']+df['SibSp']
```

```python
df.head(5)
```

```python
# Now we will
def family_type(number):
    if number ==0:
        return "Alone"
    elif number >0 and number <=4:
        return "Medium"
    else:
        return "Large"
```

```python
df['Family_type']=df['Family_size'].apply(family_type)
```
