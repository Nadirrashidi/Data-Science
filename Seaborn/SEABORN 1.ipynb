    A scatter plot is a type of data visualization that displays individual data points along two axes, 
    typically the x-axis and the y-axis. Each point represents the value of two variables. Scatter plots are useful for

The hue parameter in Seaborn's relplot() function allows you to add an additional categorical variable to the plot, 
where different levels of this variable are distinguished by different colors. This is useful for visualizing relationships
between variables while also considering the effect of a third categorical variable.
Here's an example of how to use the hue parameter in a line plot:

```python
sns.relplot(x="total_bill", y="tip",hue='smoker',data=data)
```

```python
sns.relplot(x="total_bill", y="tip",hue='sex',data=data)
```

Categorical scatterplots(catplot) :Categorical scatterplots are a type of plot used to visualize the relationship between two 
categorical variables. Unlike traditional scatterplots, which typically display the relationship between two continuous 
variables, categorical scatterplots deal with discrete categories.
In a categorical scatterplot:
Each data point represents a combination of categories from the two variables being plotted.
The points are distributed across the plot based on the categories they belong to.
It allows for the exploration of relationships between categorical variables, including identifying patterns, 
clusters, or outliers.


1.stripplot :A strip plot is a type of categorical plot used to visualize the distribution of a continuous variable within 
different categories. It displays individual data points along a single axis, with each point representing one observation in
the dataset.

2.A swarm plot is a type of categorical scatterplot that displays the distribution of data points along the categorical
axis while preventing them from overlapping. It achieves this by adjusting the positions of the points so that they do
not overlap with each other within each category.

Categorical scatterplots 
x axis pe categorical data
y axis pe numerical data

making it useful for comparing distributions across groups.

A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. 
Data type: Heatmaps are best suited for continuous or categorical data with clear values or intensities.

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# hum chahte hai ke 1949 kitne number of passengers ne travel kiya 
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,linewidth=0.5)
```

```python
sns.heatmap(x,linewidth=0.5,annot=True,fmt='d') 
```

```python
sns.heatmap(x,linewidth=0.5,annot=True,fmt='d',cmap='viridis') 
```

```python
sns.heatmap(x,linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

```python
sns.heatmap(x,linewidth=0.5,annot=True,fmt='d',cmap='Oranges')
```

The clustermap function in Seaborn is a convenient way to create a clustered heatmap, which not only visualizes the data
using a heatmap but also clusters both the rows and columns of the dataset based on their similarities. This can help in 
identifying patterns and relationships in the data.

```python
sns.clustermap(x,linewidth=0.5,cmap='Oranges',annot=True,fmt='d')
```
